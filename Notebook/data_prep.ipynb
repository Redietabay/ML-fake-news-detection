### Text preprocessing
Convert all text to lowercase

Remove punctuation, numbers, or unnecessary symbols

```python
text_cols = ['title', 'text']  


for col in text_cols:
    data[col] = data[col].str.lower()


data.head()
```

#### Remove unwanted elements

```python
import re

def clean_text(text):
    # Remove punctuation and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply to both title and text columns
data['title'] = data['title'].apply(clean_text)
data['text'] = data['text'].apply(clean_text)
```

### Remove stopwords

Stopwords are common words like “the, is, and” which usually don’t help the model because they do not give meaning  Why we do this Removing stopwords improves:
##### ✔ text cleaning
##### ✔ machine learning model accuracy
##### ✔ reduces noise
##### ✔ reduces feature size 

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in ENGLISH_STOP_WORDS])

data['title'] = data['title'].apply(remove_stopwords)
data['text'] = data['text'].apply(remove_stopwords)
```

#### Stemming or lemmatization 
it will conver words to root form

```python
from nltk.stem import PorterStemmer
ps =PorterStemmer()
def stem_text(text):
    return ' '.join([ps.stem(word) for word in text.split()])
data['title'] = data['title'].apply(stem_text)
data['text'] = data['text'].apply(stem_text)
data.head() 
```

### Combine title and text
Most Fake News models work better if you combine the title and text into one column:
WHY combine?  Because:

Title is short but very powerful (high meaning)

Text has more details
 When we combine, the model learns both

If separate → the model misses some meaning.

If combined → more context, better accuracy.

```python
data['content'] = data['title'] + ' ' + data['text']
```

#### Remove title and text

We combined the title and text into one column called **content** so the model can use all the information together. After combining, we removed the original title and text columns to avoid duplication and save memory. Using only one column also makes the model faster and easier to train. If we keep the old columns, nothing breaks, but the data becomes bigger and may reduce model accuracy.


```python
data = data.drop(['title', 'text'], axis=1)
data.head()
```

```python
if 'date' in data.columns:
    data.drop(columns=['date'], inplace=True)
```

The date column was removed because it is not useful for classification and only adds unnecessary data.Therefore, the date column was dropped to simplify the data and reduce memory usage.
